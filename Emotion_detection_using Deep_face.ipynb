```python
from deepface import DeepFace
```

```python
def emotion_analysis(emotions):
    import numpy as np
    import matplotlib.pyplot as plt
    plt.style.use('fivethirtyeight')
    objects = ('angry', 'disgust', 'fear', 'happy', 'sad', 'surprise', 'neutral')
    y_pos = np.arange(len(objects))
    plt.bar(y_pos, emotions, align='center', alpha=0.6)
    plt.xticks(y_pos, objects)
    plt.ylabel('percentage')
    plt.title('emotion')
    plt.show()
```

```python
def image_capture():
    import cv2
    cap=cv2.VideoCapture(0)
    face_cascade=cv2.CascadeClassifier('C:\\Users\\Vishal\\cv2\\data\\haarcascade_frontalface_default.xml')
    while True:
        ret,frame=cap.read()
        gray=cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
        face=face_cascade.detectMultiScale(gray,1.05,5)
        for (x,y,w,h) in face:
            cv2.rectangle(frame,(x-50,y-50),(x+w+50,y+h+50),(0,255,0),2)
            cv2.imwrite("C:\\Users\\Vishal\\datasets\\face_exp.jpg",gray[y-50:y+h+50,x-50:x+w+50])
        cv2.imshow('video',frame)
        key=cv2.waitKey(1)
        if key ==ord('v'): 
            break 
    cap.release()
    cv2.destroyAllWindows()
```

```python
def show_image_analysis():
    from keras.preprocessing import image
    import pandas as pd
    import matplotlib.pyplot as plt
    import numpy as np
    file = "C:\\Users\\Vishal\\datasets\\face_exp.jpg"
    result=DeepFace.analyze(file,actions=['emotion'])
    emotion_analysis(result['emotion'].values())
    img=plt.imread(file)
    plt.imshow(img)
    plt.title(result['dominant_emotion'])
    plt.gray()
    #print(pd.DataFrame(result['emotion']))
```
